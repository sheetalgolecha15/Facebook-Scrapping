```python
pagesrc = driver.page_source
soup = BeautifulSoup(pagesrc)
contentpage = soup.find_all('div',{'class':'_2eec'})
```

```python
print(contentpage)
```

```python
image = contentpage[0].find_all('img', {"class":'img', 'height':200})

```

```python

for i in range(len(image)):
    r = requests.get(image[i]['src']) # create HTTP response object

    filename = '/Users/sheetalgolecha/Documents/nike/test_low/' + str(i) + '.png'
    with open(filename,'wb') as f:
        f.write(r.content)
        f.close()
    
```

```python
high_res= contentpage[0].find_all('a',{'rel':'theater'})
```

```python
for i in range(len(high_res)):
    r = requests.get(high_res[i]['href']) # create HTTP response object

    filename = '/Users/sheetalgolecha/Documents/nike/high_res/' + str(i) + '.png'
    with open(filename,'wb') as f:
        f.write(r.content)
        f.close()
```
